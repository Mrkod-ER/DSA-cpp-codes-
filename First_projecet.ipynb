# My first project


# **Load data**


```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv")
df
```

# **Data prepration**


## Data sepration into x and y


```python
y = df['logS']
y
```

```python
x = df.drop('logS', axis = 1)
x
```

## **Data spliting**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 100)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

# **Model Building**

## **Linear Regression**

### **Training the model**

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

### **Apply the model to make the prediction**

```python
y_train_lr_pred = lr.predict(x_train)
y_test_lr_pred = lr.predict(x_test)
```

```python
y_train_lr_pred
```

```python
y_test_lr_pred
```

### **Evaluate the Model Prediction**

```python
from sklearn.metrics import mean_squared_error, r2_score
lr_train_mse = mean_squared_error(y_train, y_train_lr_pred)
lr_train_r2 = r2_score(y_train, y_train_lr_pred)

lr_test_mse = mean_squared_error(y_test, y_test_lr_pred)
lr_test_r2 = r2_score(y_test, y_test_lr_pred)
```

```python
print('LR MSE Train ', lr_train_mse)
print('LR R2 Train ', lr_train_r2)
print('LR MSE Test ', lr_test_mse)
print('LR R2 Test ', lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear Regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Methods', 'Train Mse', 'Train R2', 'Test Mse', 'Test R2']
```

```python
lr_results
```

## **Random Forest**

###**Training the model**

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth =2, random_state=100)
rf.fit(x_train, y_train)
```

### **Apply the model to make prediction**

```python
y_train_rf_pred = rf.predict(x_train)
y_test_rf_pred = rf.predict(x_test)
```

### **Evaluate the model Prediction**

```python
from sklearn.metrics import mean_squared_error, r2_score
rf_train_mse = mean_squared_error(y_train, y_train_rf_pred)
rf_train_r2 = r2_score(y_train, y_train_rf_pred)

rf_test_mse = mean_squared_error(y_test, y_test_rf_pred)
rf_test_r2 = r2_score(y_test, y_test_rf_pred)
```

```python
rf_results = pd.DataFrame(['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Methods', 'Train Mse', 'Train R2', 'Test Mse', 'Test R2']
```

```python
rf_results
```

# Model comparison

```python
df_models = pd.concat([lr_results, rf_results], axis=0)
df_models.reset_index(drop=True)
```
